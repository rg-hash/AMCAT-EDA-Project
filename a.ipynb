```python
##detect outliers
# Function to detect outliers using Z-score
def detect_outliers_zscore(column, threshold=3):
    z_scores = (column - column.mean()) / column.std()
    outliers = column[abs(z_scores) > threshold]
    return outliers

# Detect outliers in each column of numerical
numerical_columns = df.select_dtypes(include=['number']).columns
outliers_zscore = {}
for column in numerical_columns:
    outliers_zscore[column] = detect_outliers_zscore(df[column])

for column, outliers in outliers_zscore.items():
    print("Column '{}':".format(column))
    print(outliers)    
```

```python
import matplotlib.pyplot as plt
descriptive_stats = df[numerical_columns].describe()

# Calculate skewness and kurtosis
skewness = df[numerical_columns].skew()
kurtosis = df[numerical_columns].kurtosis()

# Plot histograms for each numerical column
for column in numerical_columns:
    plt.figure(figsize=(8, 6))
    plt.hist(df[column], bins=20, color='skyblue', edgecolor='black')
    plt.title('Histogram of {}'.format(column))
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

# Print descriptive statistics
print("Descriptive Statistics:")
print(descriptive_stats)

# Print skewness
print("\nSkewness:")
print(skewness)

# Print kurtosis
print("\nKurtosis:")
print(kurtosis)
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns

# Calculate frequency distribution for each categorical column
frequency_distribution = {}

for column in categorical_columns:
    frequency_distribution[column] = df[column].value_counts()

# Plot bar plots for each categorical column
for column in categorical_columns:
    plt.figure(figsize=(8, 6))
    frequency_distribution[column].plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title('Frequency distribution of {}'.format(column))
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

# Print frequency distribution
print("Frequency Distribution:")
for column, distribution in frequency_distribution.items():
    print("\nColumn '{}':".format(column))
    print(distribution)
```

```python

contingency_table = pd.crosstab(df['Gender'], df['Specialization'])

from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")

# Calculate Cramer's V
n = contingency_table.sum().sum()
cramer_v = (chi2 / n) ** 0.5
print(f"Cramer's V: {cramer_v}")
```
